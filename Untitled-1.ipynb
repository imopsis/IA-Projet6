```python
import spacy
print('------------------------------------DEBUT---------------------------------------------------------------')
# Load the Spacy model
nlp = spacy.load("en_core_web_sm")

# And parse the sentence (taken from which movie?)
tokens = nlp(' '.join(corpus))
print('-------------------------------------NLP TOKENS--------------------------------------------------------------')
#print(tokens)

from nltk import ngrams
from nltk.tokenize import WordPunctTokenizer

# Tokenize
tokens = WordPunctTokenizer().tokenize(' '.join(corpus))
print('---------------------------------NLTK TOKEN------------------------------------------------------------------')
#print(tokens)

from nltk.corpus import stopwords
stopwords = stopwords.words("english")

new_stopword = ["_", ".", ",", "?", "!", "-", ";"]
stopwords.extend(new_stopword)

tokens = [tk for tk in tokens if tk not in stopwords]

# Only keep the bigrams 
bigrams = [w for w in  ngrams(tokens,n=2)]

bi_tokens = ['_'.join(w) for w in bigrams]
print('------------------------------------------BI TOKEN FROM NLTK---------------------------------------------------------')
print(bi_tokens)
```

```python

from nltk.tokenize import WordPunctTokenizer
from collections import Counter

word_counts_bi = Counter(bi_tokens)
print(len(word_counts_bi))

for w in word_counts_bi.most_common(20):
   print(f"{w[0]}:  \t{w[1]} ")
```

```python
from textblob import TextBlob
text = ' '.join(corpus)
blob = TextBlob(text)

# store the results in a list of dictionaries
df = []
for sentence in blob.sentences:
    df.append({
        'verse': sentence.raw,
        'score': sentence.sentiment.polarity
    })
# convert the result into a dataframe
df = pd.DataFrame(df).sort_values(by = 'score').drop_duplicates()
# 5 first rows
df
```
