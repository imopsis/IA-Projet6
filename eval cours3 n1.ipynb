```python
from nltk.corpus import stopwords

from nltk.tokenize import WordPunctTokenizer

# Tokenize as usual
tokens  = WordPunctTokenizer().tokenize(text)

# Filter out stopwords
tokens = [tk for tk in tokens if tk not in stopwords.words('english')]

# and count the words
word_counts = Counter(tokens)
print(len(word_counts))
for w in word_counts.most_common(20):
   print(f"{w[0]}:  \t{w[1]} ")
```

```python
from nltk.stem import PorterStemmer

# Now, instantiate a stemmer
ps = PorterStemmer()

# and stem the tokens
stems   = [ps.stem(tk) for tk in tokens ]

word_counts = Counter(stems)
print(len(word_counts))
for w in word_counts.most_common(20):
   print(f"{w[0]}:  \t{w[1]} ")
```

```python
import re

def split_text_by_characters(text, char_list):
    # CrÃ©e une expression rÃ©guliÃ¨re avec les caractÃ¨res de la liste
    regex_pattern = '[' + ''.join(re.escape(char) for char in char_list) + ']'
    
    # Utilise re.split pour dÃ©couper le texte Ã  chaque caractÃ¨re de la liste
    split_text = re.split(regex_pattern, text)
    
    # Exclut les caractÃ¨res de la liste du rÃ©sultat
    #result = [token for token in split_text if token not in char_list and token]

    # Exclut les phrases vides ou celles contenant uniquement un espace
    result = [sentence.strip() for sentence in split_text if sentence.strip() and sentence != ' ']
    
    
    return result
```

```python
def count_words_in_sentences(sentences):
    # Compte le nombre de mots dans chaque phrase
    word_counts = [len(sentence.split()) for sentence in sentences]
    return word_counts

def calculate_average_word_count(word_counts):
    # Calcule la moyenne du nombre de mots par phrase
    if len(word_counts) == 0:
        return 0
    average = sum(word_counts) / len(word_counts)
    return average
```

```python
# Compter les mots dans chaque phrase
word_counts = count_words_in_sentences(phrase)

# Calculer la moyenne du nombre de mots par phrase
average_word_count = calculate_average_word_count(word_counts)

print(f"Sentences: {phrase}")
print(f"Word counts: {word_counts}")
print(f"Average word count: {average_word_count:.2f}")
```
