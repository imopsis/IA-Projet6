```python
def Get_resto_yelp(ville, offset, limit, key):
    url = 'https://api.yelp.com/v3/businesses/search'
    headers = {
        'accept' : 'application/json',
        'Authorization' : f'Bearer {key}'
        }
    parameters ={
        'location' : ville,
        'term' : 'restaurants',
        'limit' : int(limit),
        'offset' : int(offset)
    }

    response = requests.get(url, headers=headers, params=parameters)
    data = response.json()
    print(data)
    ids = []
    names = []
    ratings = []
    pricing = []
    reviews_count = []
    is_closed = []

    if 'businesses' in data:
        for business in data['businesses']:
            
            ids.append(business.get('id', 'N/A'))
            names.append(business.get('name', 'N/A'))
            reviews_count.append(business.get('review_count', 'N/A'))
            ratings.append(business.get('rating', 'N/A'))
            is_closed.append(business.get('is_closed', 'N/A'))
            try:
                pricing.append(len(business.get('price', '')))
            except:
                pricing.append(None)

    resto = {'id' : ids, 'name': names, 'rating': ratings, 'price': pricing, 'reviews_count' : reviews_count, 'is_closed' : is_closed}
    return pd.DataFrame(resto)
```

```python
import requests

def get_reviews_yelp(business_id, key):
    #url = f'https://api.yelp.com/v3/businesses/{business_id}/reviews'
    url = f"https://api.yelp.com/v3/businesses/{business_id}/reviews?limit=50&sort_by=yelp_sort"
    headers = {
        'Authorization': f'Bearer {key}'
    }

    response = requests.get(url, headers=headers)
    data = response.json()

    reviews = []
    if 'reviews' in data:
        for review in data['reviews']:
            reviews.append({
                'id' : review['id'],
                'user': review['user']['name'],
                'rating': review['rating'],
                'text': review['text']
            })

    return reviews

i = 0
j = 0
tour = 0
reviews_for_dataframe = []
while j < 200:
    df = Get_resto_yelp("Paris", tour, 50, key)
    for index, row in df.iterrows():
        print(f"len: {len(df)}")
        business_id = row['id']
#        print(f"***************** Nom du resto: {row['name']} *****************")
        reviews = get_reviews_yelp(business_id, key)


        for review in reviews:
            i += 1
            if review['rating'] > 2:
                continue
            j += 1
#            print(f"i: {i}")
#            print(f"Id: {review['id']}")
#            print(f"User: {review['user']}")
#            print(f"Rating: {review['rating']}")
#            print(f"Review: {review['text']}")
            review_for_dataframe = {'review_id' : review['id'], 'business_id' : business_id, 'stars	' : review['rating'], 'text' : review['text']}
            reviews_for_dataframe.append(review_for_dataframe)
    tour += 50

dg = pd.DataFrame(reviews_for_dataframe)
print("***************** FIN *****************")
print(f"i: {i}")
print(f"j: {j}")
print(f"tour: {tour}")
```
